#  Animated Scatterplot Exercise (Series of png files are generated)

```python
for x in range(1960,2016):
    #sns.set()
    fig, ax = plt.subplots(figsize=(9, 7))

    sns.set_style("whitegrid")
    #
    df_yr1960_2016 = df[df['year']==x]
    fig =sns.scatterplot(x='fertility_rate', y='life_expectancy', hue='country', size='population', 
    data= df_yr1960_2016, alpha=1.0, sizes=(20, 500),legend=False, cmap="inferno")

    plt.legend(bbox_to_anchor=(0, 1.5),borderaxespad=0)
    
         
 
     
    fig.set_title( x, fontsize=25)
    fig.set_ylabel("Life expectancy in years from birth",fontsize=20)
    fig.set_xlabel("Fertility rate per woman in birth",fontsize=20)

    fig.tick_params(labelsize=15)
    #plt.tight_layout()

    fig.set_xlim(0, 9)
    fig.set_ylim(10,90)
#    plt.show()
    filename = 'lifeexp_{}.png'.format(x)
    plt.savefig(fname=filename, dpi=200)

    plt.close()
```

##  Animated Scatterplot Exercise (generated  png files are combined as output.gif)

```python
import imageio

images = []
for i in range(1960,2016):
 
    filename = 'lifeexp_{}.png'.format(i)
    images.append(imageio.imread(filename))
    
imageio.mimsave('output.gif', images, fps=20)
```

#  Animated Scatterplot Exercise (The video is embedded using Celluloid Matplotlib Animations)

#### https://towardsdatascience.com/the-simplest-way-of-making-gifs-and-math-videos-with-python-aec41da74c6e

```python
from celluloid import Camera # getting the camera
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import HTML
fig, ax = plt.subplots(figsize=(9, 7)) # let's make it bigger
camera = Camera(fig)# the camera gets our figure



for x in range(1960,2016):


    #sns.set()
    sns.set_style("whitegrid")
    #
    df_yr1960_2016 = df[df['year']==x]

    

    fig =sns.scatterplot(x='fertility_rate', y='life_expectancy', hue='country', size='population', 
    data= df_yr1960_2016, alpha=1.0, sizes=(20, 500),legend=False, cmap="inferno")
    


    plt.legend(bbox_to_anchor=(0, 1.5),borderaxespad=0)
    
#    fig.set_title( x, fontsize=15) 

    fig.set_ylabel("Life expectancy in years from birth",fontsize=20)
    fig.set_xlabel("Fertility rate per woman in birth",fontsize=20)
    
    fig.tick_params(labelsize=15)
    #plt.tight_layout()

    fig.set_xlim(0, 9)
    fig.set_ylim(10,90)
#    plt.show()    
    
     
          
    ax.text(0.1, 1.01, "Life expectancy vs Fertility rate for all countries = "+str(x), 
    transform=ax.transAxes, fontsize=15) # making the dynamic title
    camera.snap()

    filename = 'lifeexp_{}.png'.format(x)

    
animation = camera.animate()
HTML(animation.to_html5_video())

 


 
```
